### Takeaway 2: This dataset records the greenhouse gas emissions of a variety of industries, with power plants, oil and gas, and waste facilities dominating.
The dataset records the emissions of nearly 6000 facilities in the U.S., representing 66 industries. The most prevalent industries recorded are power plants (8%), oil and gas (8%), and waste facilities (8%).

### Takeaway 3: The vast majority of facilities fell into the lowest emissions category.
5826 facilities emit 0-900,000 metric tons of carbon dioxide equivalent, and only between 1 and 300 facilities each emit in the higher-polluting categories.


### Takeaway 4: Only 7.7% of facilities report deploying continuous emissions monitoring.
The GHG Reporting Program allows faciltiies to estimate emissions using emissions factors and extrapolations from engineering estimates, rather than requires them to deploy continuous emission monitoring. However, power plants, which constitute about 8% of the dataset, are requried to have continous emissions monitoring per the Clean Air Act. Thus, those that deploy continuous monitoring do so eiteher voluntarily or because they are power plants regulated under the CAA.
